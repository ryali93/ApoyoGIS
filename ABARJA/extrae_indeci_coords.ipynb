```python
import requests
from bs4 import BeautifulSoup
```

```python
def extrae_coord(codigo):
    url = 'http://sinpad.indeci.gob.pe/Sinpad/emergencias/Evaluacion/Reporte/rpt_eme_situacion_emergencia.asp?EmergCode={}'.format(codigo)
    s = requests.get(url)
    html = s.text
    soup = BeautifulSoup(html, 'html.parser')
    tds = []
    for s in soup.find_all('td'):
        try:
            if s != None:
                texto = s.get_text(strip=True).replace(",", ".")
                num = float(texto)
                if num < 0:
                    tds.append(num)
        except:
            pass
    return tds
```

```python
import pandas as pd
```

```python
df = pd.read_csv("657185058.csv", encoding = "cp1252")
df["Codigo"] = df["Codigo"].apply(lambda x: str(x).zfill(8))
df.head()
```

```python
df["XY"] = df["Codigo"].apply(extrae_coord)
```

```python
df["X"] = df["XY"].apply(lambda x: x[0])
df["Y"] = df["XY"].apply(lambda x: x[1])
```
